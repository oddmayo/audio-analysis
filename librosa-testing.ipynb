## Autocorrelation

```python
import matplotlib.pyplot as plt
odf = librosa.onset.onset_strength(y=y, sr=sr, hop_length=512)
ac = librosa.autocorrelate(odf, max_size=4 * sr // 512)
fig, ax = plt.subplots()
ax.plot(ac)
ax.set(title='Auto-correlation', xlabel='Lag (frames)')
```

```python
odf = librosa.onset.onset_strength(y=y2, sr=sr2, hop_length=512)
ac = librosa.autocorrelate(odf, max_size=4 * sr2 // 512)
fig, ax = plt.subplots()
ax.plot(ac)
ax.set(title='Auto-correlation', xlabel='Lag (frames)')
```

```python
# compute autocorrelations (keeps variables distinct)
hop_length = 512

odf1 = librosa.onset.onset_strength(y=y,  sr=sr,  hop_length=hop_length)
ac1  = librosa.autocorrelate(odf1, max_size=4 * sr // hop_length)

odf2 = librosa.onset.onset_strength(y=y2, sr=sr2, hop_length=hop_length)
ac2  = librosa.autocorrelate(odf2, max_size=4 * sr2 // hop_length)

# plot side-by-side
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 4), sharey=True)
axes[0].plot(ac1)
axes[0].set(title='Autocorrelation — file A', xlabel='Lag (frames)')
axes[1].plot(ac2)
axes[1].set(title='Autocorrelation — file B', xlabel='Lag (frames)')

fig.tight_layout()
```

The x‑axis is lag in frames: each point is how many analysis frames the signal is time‑shifted before correlating with itself; convert frames to seconds with time = frames × hop_length / sr, or map to BPM with tempo_frequencies if analyzing tempo.​

The y‑axis is autocorrelation amplitude: the similarity (correlation) between the signal and its time‑shifted copy at that lag; it peaks at zero lag and reflects repeated patterns at non‑zero lags.​

The plot shows a strong peak at zero lag and several regularly spaced secondary peaks, indicating clear periodicity in the audio and suggesting a stable tempo or repeating structure in the song’s onset/energy pattern.[1][2]

### What the peaks mean
- Autocorrelation is maximal at lag 0 and symmetric; secondary peaks at non-zero lags occur when the signal correlates with a time-shifted version of itself, revealing repeated patterns such as pitch periods (short lags) or tempo pulses (longer lags).[3][2]
- When there is a prominent periodic component, peaks appear at the fundamental lag and at integer multiples of that lag; evenly spaced peaks across the lag axis typically indicate a consistent beat period in music.[4][1]

### Reading this specific plot
- The tall spike at lag 0 is expected for all signals; what matters here are the recurring peaks at roughly equal frame intervals, which imply a regular repetition rate consistent with a steady tempo in the underlying song segment.[2][1]
- If those secondary peaks are strong relative to the baseline, the onset/novelty function or energy is repeating reliably; their spacing can be converted to tempo by mapping lag (in frames) to seconds and then to beats per minute $$ \text{BPM} = 60 / \text{period in seconds} $$.[4][2]

### How it can differ across songs
- Songs with steady drums or metronomic patterns produce clear, high secondary peaks at uniform lags; songs with rubato, complex rhythms, or weak percussive onsets yield broader, weaker, or irregular peaks, making tempo less certain.[5][6]
- Highly periodic pitched tones show strong short-lag peaks due to the fundamental frequency, while noisy or inharmonic textures have flatter short-lag regions; preprocessing or using onset novelty before autocorrelation mitigates timbral confounds when focusing on tempo.[7][1]

### Practical notes
- Estimating tempo often analyzes the largest non-zero-lag peaks (and their multiplicity) to pick the beat period, sometimes checking the top few peaks to avoid octave errors (e.g., half/double tempo).[8][4]
- Peak salience can be affected by non-tempo structures; techniques like center clipping or difference-function methods (e.g., YIN) reduce spurious peaks when estimating periodicity from raw audio.[9][7]

### Bottom line
- Compared to arbitrary audio, this output looks like music with a stable repetitive pattern: strong, regularly spaced non-zero-lag peaks suggest a consistent tempo; other songs with less regular rhythm or more expressive timing would show weaker or less evenly spaced peaks.[6][1]

### Song B (bottom plot)
